```python
#Import libraries
import pandas as pd #https://pandas.pydata.org/
import matplotlib.pyplot as plt #https://matplotlib.org/
import matplotlib.dates as mdates #https://matplotlib.org/
import numpy as np #http://www.numpy.org/


Temperature = pd.read_csv("temp-2020-2021-LPR-airport.csv",dayfirst=True,sep=",",
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])
Temperature.info()
```

```python
#plotting one year
plt.figure(figsize=(16,6))
plt.plot(Temperature,color='blue', marker='x',linestyle=':')
plt.title("Air temperature for one year 2020-2021, at Lappeenranta airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()

#plotting one month
plt.figure(figsize=(16,6))
plt.plot(Temperature['2020-9':'2020-9-30'],color='blue', marker='x',linestyle=':')
plt.title("Air temperature for September 2020, at Lappeenranta airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()

#plotting one week (my arrival week in Lappeenranta!)
plt.figure(figsize=(16,6))
plt.plot(Temperature['2020-8-16':'2020-8-22'],color='blue', marker='x',linestyle=':')
plt.title("Air temperature for one week in August 2020, at Lappeenranta airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()

#plotting one day
plt.figure(figsize=(16,6))
plt.plot(Temperature['2020-12-16'] ,color='black', marker='o',linestyle=':')
plt.title("Air temperature in December 16, 2020 (my birthday!), at Lappeenranta airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()
```

**(4) Consider a house with an electric heater. How the electricity consumption of the heater would be affected by the air temperature? Is there a way to change this?**

Answer:

The electricity consumption would be affected by the ambient air temperature because the resistance of materials changes as the
temperature of the material changes. Increasing the temperature usually increases the resistance which will generate more heat.
So when the air temperature is cold the heater will not have as much resistance as when it is warm so the same amount of energy will
produce less heat. Or conversely to generate an equivalent amount of heat you will need more electricity. Additionally if the heater is
equipment with a temperature measurement that will turn it off when it reaches a desired temperature this will dramatically impact
energy consumption since if the air temperature is already close to the desired temperature the heater will run much less than if the
air temperature was drastically colder than the setpoint of the heater. In this case strong insulation on the house will retain heat longer
and cause the heater to run less therefore using less energy. Also placing the heater in the center of the room will dissipate the heat more evenly
and lead to more comfort and more efficient use of electricity.
